```python
import numpy as np
import matplotlib.pyplot as plt

def function(x):
    fx = (x+3)*(x+2)*x*(x-2)*(x-3)
    return fx

x = np.linspace(-3.4,3.4)
y = function(x)

plt.plot(x, y, color='blue', linewidth=2, label='this function')
plt.xlim([-5,5]) 
plt.ylim([-100, 100])    
plt.xlabel('x')          
plt.ylabel('f(x)')     
plt.title('Fyrirlestraræfing 4') 
plt.axhline(0, color='black',linewidth=0.5)  
plt.axvline(0, color='black',linewidth=0.5) 


plt.tick_params(axis='both',          # Apply to both x and y axes
                which='major',        # Apply to major ticks
                direction='in',       # Ticks inside the plot
                length=8,             # Length of ticks
                width=1,              # Width of ticks
                color='yellow',        # Color of ticks
                pad=8,                # Distance between ticks and labels
                labelsize='large',    # Size of the tick labels
                labelcolor='purple') 

# Show the plot
plt.show()
```
