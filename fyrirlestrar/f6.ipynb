```python
import pandas as pd
from urllib.request import urlopen
kosningar = urlopen("https://cs.hi.is/python/kosningar-2021.txt")
flokk = urlopen("https://cs.hi.is/python/flokksnofn.txt")

df1 = pd.read_fwf(kosningar)
df2 = pd.read_csv(flokk, sep='\t')
samtals = sum(df1["Atkvæði"])
hlutfall = (df1["Atkvæði"] / samtals) * 100
df1["Hlutfall"] = hlutfall
joined = df1.join(df2)
display(joined)

df3= df1[df1['Hlutfall'] > 5] 
display(df3)
pd.options.display.float_format = '{:.0f}%'.format 







```
