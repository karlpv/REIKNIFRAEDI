```python
import math 
x = 2 
y = 2
f = 3*math.pow(x,2) + math.sqrt(math.pow(x,2)+math.pow(y,2)) + math.exp(math.log(x))
print(f)
```

The same code, but in a function

```python
def funct(x,y):
    f = 3*math.pow(x,2) + math.sqrt(math.pow(x,2)+math.pow(y,2)) + math.exp(math.log(x))
    return f

x = 2
y = 2

print(funct(x,y))
    
```
