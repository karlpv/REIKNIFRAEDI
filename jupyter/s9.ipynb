# VV7. Lágmörkun Rosenbrock-fallsins


```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as opt

def rosen(x):
  res = (1-x[0])**2 + 100*(x[1] - x[0]**2)**2
  return res

def cb(xk):
    plt.scatter(xk[0], xk[1])
    
def rosg(x):
  fx0 = -2*(1-x[0]) - (200*(x[1]-x[0]**2)*(2*x[0]))
  fx1 = 0 + 200*(x[1]-x[0]**2)*(1)
  gradient = np.array([fx0,fx1])
  return gradient

plt.figure(figsize=(7,7))
x = np.linspace(-1.5,1.5,400)
y = np.linspace(-0.5,2.5,400)
[X,Y] = np.meshgrid(x,y)
Z = rosen([X,Y])
levels = [0.2, 5, 50] + list(range(200,2300,400))
c = plt.contour(X, Y, Z, levels = levels, colors = 'gray')
levstr = {l:str(l) for l in levels}
plt.clabel(c, fmt=levstr)
plt.xticks(range(-2,4))
plt.yticks(range(-1,5))
plt.grid("True")


x0 =(-1.2,1)
xstar=(1,1)
gildi_x0 = rosen(x0)
gildi_xstar = rosen(xstar)
print(f'gildi x0:{gildi_x0:.1f}')
print(f'gildi x*:{gildi_xstar}')


result = opt.minimize(rosen,x0,callback=cb,method='BFGS',jac=rosg)
xmin = result.x
print(result)
print(xmin) 

print(rosg(x0))
print(rosg(xstar))





```
